```python
import os
import pandas as pd
import numpy as np
import pathlib
import xarray as xr
import rioxarray
```

```python
read_file_path: str = r"./../data_global_surge/field_2024-09-22_12_00_00.f0.nc"
"""读取文件"""
data = xr.open_dataset(read_file_path)
data
```

```python
data_standard=data.sortby('latitude',ascending=False)
```

```python
# 将dataset中的默认值替换为nan
DEFAULT_VAL=0.0
data['h']
```

```python
first_time = data_standard["Time"].values[0]
```

```python
first_dr = data_standard.sel(Time=first_time)["h"]
# 设置空间坐标系与分辨率
first_dr.rio.set_spatial_dims(x_dim="longitude", y_dim="latitude", inplace=True)
first_dr.rio.write_crs("EPSG:4326", inplace=True)
```

```python
out_put_file_path: str = (
    r"./../data_global_surge/field_2024-09-22_12_00_00.f0_jpeg.tif"
)
# first_dr.rio.to_raster(out_put_file_path,diver='GTiff',compress='deflate')
first_dr.rio.to_raster(out_put_file_path,diver='GTiff',compress='jpeg')
```
